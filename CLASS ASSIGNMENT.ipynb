```python


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"{self.name} has enrolled in {course_name}.")

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")


class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"{self.name} is teaching {course_name}.")


S1 = Student("Emmanuel", 20, "ST202")
L1 = Lecturer("Dr. Kofi Boateng", 45, "EMP202")

S1.display_info()
L1.display_info()
S1.enroll_course("Engineering")
L1.teach_course("Programming")
```
